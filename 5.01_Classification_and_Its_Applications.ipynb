##__1. Classification__##

Classification is a supervised machine learning technique where the model is trained to predict the class label of a given input data. It looks for the decision boundary, which divides the dataset into different classes.


Classification algorithms classify or categorize data into a class or category.

###__3.1 Binary Classification__###

 Binary classification is a type of supervised learning task in machine learning where the goal is to categorize data into one of two possible classes or categories.
These classes are often denoted as positive (1) and negative (0).

Example: Predicting whether an apple is rotten has only two possibilities.

###__3.2 Ordinal Classification__###

Ordinal classification is a classification task where the target variable has ordered categories or levels.



Example: Predicting the student performance levels (low, medium, or high) based on study hours, previous test scores, and participation level in a class

Some of the popular algorithms used for ordinal classification are:

###__3.3 Multiclass Classification__####

Multiclass classification is a supervised learning task where the goal is to categorize data instances into one of several predefined classes or categories. Unlike binary classification, which deals with only two classes, multiclass classification involves handling multiple classes.

###__3.4 Multi-Label Classification__###

It allows us to classify datasets with more than one target variable. There are several labels for the outputs of a given prediction, and when making predictions, a given input may belong to more than one label.





Example: When predicting a given movie’s category, it may belong to horror, romance, adventure, action, or all simultaneously. Here, some multi-labels can be assigned to a given movie.

###__3.5 Anomaly Detection__###


Anomaly Detection is a specialized classification algorithm that identifies rare or unusual data instances.


Example: Anomaly detection in credit card fraud detection, where the algorithm classifies transactions as normal or fraudulent based on features like transaction amount, location, and spending patterns, aiming to identify unusual transactions deviating from the expected behavior.

Some algorithms used for anomaly detection are:
- K Means Clustering
- Support Vector Machine (SVM)

###__3.6 Imbalanced Classification__###

Imbalanced classification involves an unequal distribution of classes.



The imbalance in the class distribution may vary, but a severe imbalance is more challenging to model and may require specialized techniques like:
- Random undersampling
- SMOTE oversampling


Many real-world classification problems, such as fraud detection, spam detection, and churn prediction, have an imbalanced class distribution.

##__5. Naive Bayes Classifier__##


The Naive Bayes classifier is a machine learning model that segregates different objects based on certain features or variables.


The Naive Bayes classifiers are based on Bayes' theorem. It predicts the probable occurrence of an event based on prior knowledge of other associated events.


Example: The probability of a house's price being high can be assessed with knowledge about the neighborhood.



Bayes' theorem works on the principle of contingent probability.


![link text](https://labcontent.simplicdn.net/data-content/content-assets/Data_and_AI/ML/New/Lesson_05_Supervised_Learning_Classification/bayes_theorem.png)

P(A) is the probability of the hypothesis H being true.

P(B) is the probability of the evidence being true.

P(B|A) is the probability of the evidence that the hypothesis is true.

P(A|B) is the probability of the hypothesis that the evidence is true.




It measures the probability that something will happen, given that something else has already occurred.

## Stochastic Gradient Descent ##


Stochastic Gradient Descent (SGD) is a common machine learning algorithm and forms the basis of neural networks.


The word **gradient** means a slope or slant.


image.png

- Gradient descent is a descending slope that helps reach the lowest point on a surface.
- The lowest point of the parabola occurs when x = 1.
- The gradient descent algorithm aims to find the value of x such that the value of y is minimal.
- y is the objective function that descends to the lowest point.




It is an iterative algorithm that starts from a random point on a function and travels down its slope in steps until it reaches the function's lowest point.


The steps involved in a gradient descent algorithm are:



Step 1: Compute the gradient of the function, which helps find the slope of the objective function.




Step 2: Pick a random initial value for the parameters (in this case, differentiate y with respect to x).



Step 3: Plug in the parameter values and update the function.


Step 4: Calculate the step size for each feature: Step size = gradient * learning rate.


Step 5: Calculate the new parameters: New params = old params - step size

Step 6: Repeat steps 3 to 5 till the gradient
is almost zero.

- The learning rate is a flexible parameter that influences the algorithm's convergence towards the minimum.
- The larger the learning rates, the larger the steps taken down the slope, potentially resulting in overshooting the minimum.
- This may lead to the minimum being missed.
- It is thus advisable to use small learning rates.

#### Advantages of Stochastic Gradient Descent (SGD) ####

- It offers speed, efficiency, and ease of implementation.

#### Disadvantages of Stochastic Gradient Descent (SGD) ####

- Due to frequent updates, the steps taken toward the minimum may be noisy, which may lengthen the time taken to reach the minima.
- Applying frequent updates to a large dataset is computationally expensive.

#### Applications of SGD ####

When companies find it difficult to keep a quality check, machine learning helps to automate the process.
- It helps tackle low-performance issues in the model created.
- It helps speed up convergence in huge data sets.
- It is used in financial modeling to reduce inaccuracies in predictions and improve model performance.

## Step 3: Define the CustomPerceptron Class and Its Methods

- Define `predict`, `fit`, `activation`, and `score` methods for the CustomPerceptron class.

```python
class CustomPerceptron(object):

    def __init__(self, n_iterations=100, random_state=1, learning_rate=0.01):
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.learning_rate = learning_rate

    '''
    Stochastic Gradient Descent

    1. Weights are updated based on each training examples.
    2. Learning of weights can continue for multiple iterations.
    3. Learning rate needs to be defined.
    '''
    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.coef_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        for _ in range(self.n_iterations):
            for xi, expected_value in zip(X, y):
                predicted_value = self.predict(xi)
                self.coef_[1:] += self.learning_rate * (expected_value - predicted_value) * xi
                self.coef_[0] += self.learning_rate * (expected_value - predicted_value) * 1

    '''
    Activation function calculates the value of weighted sum of input value
    '''
    def activation(self, X):
            return np.dot(X, self.coef_[1:]) + self.coef_[0]

    '''
    Prediction is made on the basis of unit step function.
    '''
    def predict(self, X):
        output = self.activation(X)
        return np.where(output >= 0.0, 1, 0)

    '''
    Model score is calculated based on comparison of
    expected value and predicted value.
    '''
    def score(self, X, y):
        misclassified_data_count = 0
        for xi, target in zip(X, y):
            output = self.predict(xi)
            if(target != output):
                misclassified_data_count += 1
        total_data_count = len(X)
        self.score_ = (total_data_count - misclassified_data_count)/total_data_count
        return self.score_
```

- Evaluate the model's performance on the test and training data.


```python
prcptrn = CustomPerceptron()

prcptrn.fit(X_train, y_train)

prcptrn.score(X_test, y_test), prcptrn.score(X_train, y_train)
```

### K-Nearest Neighbors(KNN) ###


The k-nearest neighbors (k-NN) algorithm is a simple instance-based machine learning algorithm used for classification and regression tasks, where the prediction for a new data point is determined by the majority class or average of the k nearest data points in the feature space.

The following diagram represents a before-and-after KNN scenario:

![link text](https://labcontent.simplicdn.net/data-content/content-assets/Data_and_AI/ML/New/Lesson_05_Supervised_Learning_Classification/beforenafterknn.png)

- KNN is a non-parametric algorithm and doesn't make assumptions about the data.
- It stores the data set during its training phase.
- It assigns new data points to their nearest category that has similar characteristics.


The following is a step-by-step KNN process:


Step 1: Load the training with test data


Step 2: Choose the value of K, which is the nearest data point (K can be any integer)

Note: The best value for K will depend on the specific dataset and the problem to be solved.


Step 3: For every point within the test data, do the following step:


*   Calculate the space between test data and coaching data with tactics such as Euclidean, Manhattan, or Hamming distance

*   Support the space value and sort it in ascending order
*   Choose the highest k rows from the sorted array
*   Assign a category to the test point that supports the most frequent class




### Advantages of KNN ###


The KNN algorithm has the following advantages:
- Robust even with noisy training data.
- Simple to implement.
- Simple and effective if the training data is large.


### Disadvantages of KNN ###

The KNN algorithm has the following disadvantages:
- The worth of K must always be determined, which is quite complex.
- It can be computationally expensive to calculate the space between the  information points for all training samples.


### Applications of KNN ###

KNN is applied in the manufacturing industry to classify manufactured goods based on:
- Size
- Grading
- Priority
- Location

- AUC values below 0.5 indicate that the model performs worse than random guessing.

- The support for each class in a classification report is the count of samples that belong to that class in the dataset used for evaluation.
- The F1 score, a metric often used in classification reports, is the harmonic mean of precision and recall, providing a single measure that balances both measures, making it useful for evaluating classifier performance when classes are imbalanced.

The F1 score is calculated using the following formula:

F1 = ( 2 X Precision X Recall ) / (Precision + Recall )

Where:

Precision is the ratio of true positives (correctly predicted positive instances) to the sum of true positives and false positives (incorrectly predicted positive instances).

Recall is the ratio of true positives to the sum of true positives and false negatives (positive instances incorrectly predicted as negative).
