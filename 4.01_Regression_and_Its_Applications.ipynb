###__2.1 Linear Regression__###

**Linear regression** refers to a supervised learning algorithm used to predict a continuous target variable by modeling its relationship with one or more independent variables through a linear equation.
- It predicts a continuous dependent variable based on an independent variable.
- It uses the least square criterion for estimation.
- It can be applied only if there is a linear relationship between the variables.



![link text](https://labcontent.simplicdn.net/data-content/content-assets/Data_and_AI/ML/Picture6.png)

In this case, the dependent variable is continuous, and the independent variables can be either continuous or discrete.
The relationship between a dependent variable (y) and one or more independent variables (x) is established using a best-fit straight line, also known as the regression line.
The equation of the linear regression line is **y = mx +c**.


There are two types of linear regression:
- Simple linear regression
- Multiple linear regression

In a simple linear regression, the value of a numerical dependent variable is predicted using a single independent variable.


image.png

In multiple linear regression, more than one independent variable is used to predict the value of a numerical dependent variable.


image.png

A few important assumptions in regression analysis are as follows:
- The error terms are normally distributed.
- The independent variables are not correlated.
- Error or residual terms are not correlated.
- The error terms have constant variance.

###__2.7 Ridge Regression__###

**Ridge regression** is a regularization technique used in linear regression to prevent overfitting of the model. It is used when dealing with multicollinear data.
- Least squares are unbiased, and variances are large.
- The predicted values will be far away from the actual values.
It adds a regularization term to the loss function that penalizes large coefficients in the model and encourages the model to have smaller coefficients.
- It results in a simpler and more generalized model.

### Hyperparameter tuning ###

- Hyperparameter tuning is the process of finding the best settings for the parameters in a machine-learning model.
- Hyperparameters are settings that are not learned during training but are set before the training process begins.
- Hyperparameter tuning involves trying different combinations of hyperparameters and evaluating the model's performance using validation techniques.

#### Step 4: Multicollinearity Check with VIF


**Multicollinearity** means that there is a strong relationship between two or more of the predictor variables you're using to predict an outcome. This can cause issues because it makes it difficult for the model to differentiate between the effects of each variable on the outcome. It can lead to unreliable estimates of the coefficients and make it hard to interpret the results accurately.

VIF stands for **Variance Inflation Factor**. It's a measure to detect multicollinearity in regression analysis. VIF tells you how much an estimated regression coefficient's variance (or uncertainty) is inflated due to multicollinearity. Higher VIF values indicate stronger multicollinearity, which can reduce the precision of coefficient estimates and complicate the interpretation of relationships between variables.

- Calculate the Variance Inflation Factor (VIF) for each predictor to identify multicollinearity.
- Consider removing or adjusting features with high VIF scores.

#### __4.1 Computing the correlation between feature variables__

```python
corr = housing_df.corr()
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
```

```python
corr["HousePrice"].abs().sort_values(ascending=False)
```

The code sorts absolute correlations between __HousePrice__ and other variables, revealing the most influential predictors. It facilitates the identification of factors strongly associated with HousePrice. For instance, __MedInc__ shows a notably high __positive correlation__ with __HousePrice__, suggesting it is a strong predictor for HousePrice prediction.

```python
# Extract 'MedInc' and 'HousePrice' columns
medinc = housing_df['MedInc']
houseprice = housing_df['HousePrice']

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(medinc, houseprice, alpha=0.5, color='blue')
plt.title('Scatter Plot of MedInc vs HousePrice')
plt.xlabel('MedInc')
plt.ylabel('HousePrice')
plt.grid(True)
plt.show()
```

- VIF quantifies how much the variance of estimated regression coefficients inflates compared to when predictor variables aren't correlated.
- A high VIF suggests multicollinearity, indicating redundancy or a strong correlation among predictors. In such cases, adjusting or removing variables may enhance model stability and interpretability.
- It also helps identify and address issues that may affect the validity and reliability of regression models.

```python
# Import the variance_inflation_factor module from the statsmodels.stats.outliers_influence library
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Set the features as the DataFrame, except the target variable
housing_df_vif = housing_df.drop('HousePrice', axis=1)
housing_df_vif = housing_df_vif.apply(pd.to_numeric, errors='coerce')

# # Calculate the variance inflation factor (VIF) for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = housing_df_vif.columns
vif_data["VIF"] = [variance_inflation_factor(housing_df_vif.values, i) for i in range(len(housing_df_vif.columns))]
print(vif_data)
```

__Observations:__

- The code calculates the Variance Inflation Factor (VIF) for each feature in the dataset, indicating the extent of multicollinearity.

- __Latitude__ and __Longitude__ exhibit high VIF scores, suggesting strong multicollinearity. Addressing these features may enhance model stability and interpretability.

```python
# Removing highly multicollinear features from the dataset
housing_df = housing_df.drop(['Latitude', 'Longitude'], axis = 1)
```

#### Step 5: Fit the Linear Regression Model
- Prepare your features (X) and target (y) for modeling.
- Add a constant to X using __sm.add_constant()__ to include an intercept term in the model.
- Fit the OLS regression model using __sm.OLS().fit()__, and print the summary to evaluate the initial model's performance.

__Note:__ The __add_constant__ function is a utility provided by statistical packages such as Statsmodels in Python to add an intercept term to your model. In linear regression, adding a constant means including a column of ones in your predictor variables (features) so that the model includes an intercept (β0) in the regression equation.

```python
# Convert selected columns to numeric

cols_to_convert = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']
housing_df[cols_to_convert] = housing_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
housing_df['HousePrice'] = pd.to_numeric(housing_df['HousePrice'], errors='coerce')

# Add a constant to include the intercept
X = sm.add_constant(housing_df.drop('HousePrice', axis=1))
y = housing_df['HousePrice']

# Fit OLS regression model
model = sm.OLS(y, X).fit()
```

#### Step 7: Model Evaluation


####Evaluation Metrics####

- Evaluation metrics help in regression analysis as they provide quantitative measures of model performance, aiding in the assessment and selection of regression models.


In regression analysis, several evaluation metrics are commonly used to assess the performance of a regression model. Some of the key evaluation metrics include:

Mean Squared Error (MSE): It calculates the average of the squares of the errors, which are the differences between the actual and predicted values. A lower MSE indicates better model performance.

Root Mean Squared Error (RMSE): RMSE is the square root of the MSE, providing a measure of the average magnitude of the errors in the predicted values. It's in the same units as the dependent variable, making it easier to interpret than MSE.

Mean Absolute Error (MAE): MAE calculates the average of the absolute errors between the actual and predicted values. It's less sensitive to outliers compared to MSE and RMSE.

R-squared (R²): R-squared measures the proportion of the variance in the dependent variable that is explained by the independent variables in the model. It ranges from 0 to 1, where a higher value indicates a better fit of the model to the data.

__Observation:__

You observe outliers in features like __MedInc,__ __Population,__ and __HousePrice__, which adversely affect model performance. It's essential to address them for improved model accuracy.

- Creating a function that iterates through each column, calculates the lower and upper bounds based on the interquartile range, and then replaces outliers with these bounds ensures that extreme values are constrained within reasonable limits.

- You substitute outliers instead of removing them to maintain dataset integrity and preserve potentially valuable insights. Outliers may contain meaningful information, but replacing them with bounds can introduce bias into statistical analyses and machine learning models, reducing their impact on the results.

```python
def remove_outliers(df):
    """
    Remove outliers from each column of a DataFrame

    Parameters:
    df (DataFrame): Input DataFrame

    Returns:
    DataFrame: DataFrame with outliers replaced by lower and upper bounds
    """
    # Create a copy of the original DataFrame to avoid modifying the original
    df_no_outliers = df.copy()

    for column_name in df.columns:
        # Calculate the upper and lower quartiles
        Q1 = df[column_name].quantile(0.25)
        Q3 = df[column_name].quantile(0.75)

        # Calculate the interquartile range
        IQR = Q3 - Q1

        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace outliers with lower and upper bounds
        df_no_outliers[column_name] = df_no_outliers[column_name].apply(lambda x: lower_bound if x < lower_bound
                                                                        else upper_bound if x > upper_bound else x)

    return df_no_outliers

housing_df_new = remove_outliers(housing_df)
```

##__4. Oversampling Using SMOTE__##



It is a dataset where the classes are not equally distributed. This can occur when one class has more instances than the other. To address the data bias between classes, oversampling can be performed using a technique called the Synthetic Minority Oversampling Technique (SMOTE).


SMOTE is a data augmentation technique that helps balance data distribution by adding to the minority class.


image.png

Working of SMOTE:

- Select an input vector from the minority class.
- Find its K-nearest neighbors​.
- Build a line joining the point under consideration
and the chosen neighbor​.
- Place a synthetic point anywhere on the line drawn.
- Repeat the steps until the data is balanced.
